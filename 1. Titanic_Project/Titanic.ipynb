```python
def CleanData(data):
    dropCols = ['Name']
    data = data.drop(dropCols, axis = 1)
    numCols = ["Age", "RoomService", "FoodCourt", "ShoppingMall", "Spa" , "VRDeck"]
    
    for col in numCols:
        data[col].fillna(data[col].median(), inplace = True)
    
    data["HomePlanet"].fillna("Earth", inplace=True)
    data["CryoSleep"].fillna(False, inplace=True)
    data["Cabin"].fillna("G/734/S", inplace=True)
    data["Destination"].fillna("TRAPPIST-1e", inplace=True)
    data["VIP"].fillna(False, inplace=True)
    
    return data
```

```python
train_c = CleanData(train)
test_c = CleanData(test)
```

```python
test_c.isna().sum()
```

```python
def Cabin_Transform(data):
    cabin = list(data)
    cabin_First = []
    cabin_Middle = []
    cabin_Last = []
    for i in cabin:
        l = str(i).split("/")
        cabin_First.append(l[0])
        cabin_Middle.append(l[1])
        cabin_Last.append(l[2])
    return cabin_First, cabin_Middle, cabin_Last
```

```python
cabin_list = [train_c['Cabin'], test_c['Cabin']]
for i in range (len(cabin_list)):
    cabin_First, cabin_Middle, cabin_Last = Cabin_Transform(cabin_list[i])
    if i==0:
        train_c.drop('Cabin', inplace = True, axis = 1)
        train_c['Cabin_First'] = cabin_First
        train_c['Cabin_Middle'] = cabin_Middle
        train_c['Cabin_Last'] = cabin_Last
    else:
        test_c.drop('Cabin', inplace = True, axis = 1)
        test_c['Cabin_First'] = cabin_First
        test_c['Cabin_Middle'] = cabin_Middle
        test_c['Cabin_Last'] = cabin_Last
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
leb_b = preprocessing.LabelBinarizer()

cols= ["HomePlanet", "Cabin_First" ,"Cabin_Middle", "Cabin_Last", "Destination", "CryoSleep", "VIP"]
for col in cols:
    train_c[col] = le.fit_transform(train_c[col])
    test_c[col] = le.fit_transform(test_c[col])
    print(f"{col} :{le.classes_}")
train_c["Transported"] = le.fit_transform(train_c["Transported"])
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = LogisticRegression(random_state=42, max_iter=5000, C=0.5).fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(preds, y_test)
```
